```python
def get_park_spreadsheet(park_chosen='Acadia National Park (ACAD)'):
    chrome_options = webdriver.chrome.options.Options()
    
    chrome_path = r'C:\Users\Sumit\Anaconda3\pkgs\python-chromedriver-binary-77.0.3865.40.0-py36_0\Lib\site-packages\chromedriver_binary\chromedriver.exe'
    driver = webdriver.Chrome(chrome_path, options=chrome_options)
    driver.get("https://irma.nps.gov/NPSpecies/Search/SpeciesList")
    # assert 'NPSpecies' in driver.title
    
    wait = WebDriverWait(driver, 15)
    driver.implicitly_wait(10)
    
    elem1 = driver.find_element_by_id("ext-gen1047").click()
    elem9 = driver.find_element_by_id("ext-gen1048").click()
    elem2 = driver.find_element_by_id("boundlist-1145-listEl") # boundlist-1145-listEl
    elem3 = elem2.find_element_by_class_name('x-list-plain')
    
    elem8 = driver.find_element_by_id('nps-npspecies-ux-filterparkcombobox-1011-inputEl')
    driver.execute_script("arguments[0].value = '"+park_chosen.replace("'","\\'")+"';", elem8)
    
    elem8.send_keys(Keys.TAB)
    elem5 = driver.find_element_by_id('radiofield-1017-inputEl').click() # Click Full list with details
    elem6 = driver.find_element_by_id('button-1021-btnIconEl').click() # Click Submit
    
    # Download:
    print(park_chosen+' downloading...')
    stale_button = True
    while (stale_button == True):
        try:
            elem12 = driver.find_element_by_id('button-1142').click()
            stale_button = False
        except:
            stale_button = True
            driver.implicitly_wait(1)
    
    driver.close()
```

```python
def get_park_spreadsheet_waits(park_chosen='Acadia National Park (ACAD)'):
    chrome_options = webdriver.chrome.options.Options()
    
    chrome_path = r'C:\Users\Sumit\Anaconda3\pkgs\python-chromedriver-binary-77.0.3865.40.0-py36_0\Lib\site-packages\chromedriver_binary\chromedriver.exe'
    driver = webdriver.Chrome(chrome_path, options=chrome_options)
    driver.get("https://irma.nps.gov/NPSpecies/Search/SpeciesList")
    # assert 'NPSpecies' in driver.title
    
    wait = WebDriverWait(driver, 15)
    driver.implicitly_wait(15)
    
    elem1 = wait.until(ec.element_to_be_clickable((By.ID, "ext-gen1047")))
    elem1.click()
    elem9 = wait.until(ec.element_to_be_clickable((By.ID, "ext-gen1048")))
    elem9.click()
    elem2 = driver.find_element_by_id("boundlist-1145-listEl") # boundlist-1145-listEl
    elem3 = wait.until(ec.presence_of_element_located((By.CLASS_NAME, 'x-list-plain')))
    # elem3 = elem2.find_element_by_class_name('x-list-plain')
    time.sleep(5)
    elem8 = driver.find_element_by_id('nps-npspecies-ux-filterparkcombobox-1011-inputEl')
    driver.execute_script("arguments[0].value = '"+park_chosen.replace("'","\\'")+"';", elem8)
    
    elem8.send_keys(Keys.TAB)
    
    stale_button = True
    while (stale_button == True):
        try:
            elem5 = wait.until(ec.element_to_be_clickable((By.ID, "radiofield-1017-inputEl")))
            elem5.click() # driver.find_element_by_id('radiofield-1017-inputEl').click() # Click Full list with details
            stale_button = False
        except:
            stale_button = True
            time.sleep(5)
    
    stale_button = True
    while (stale_button == True):
        try:
            elem6 = wait.until(ec.element_to_be_clickable((By.ID, "button-1021-btnIconEl")))
            elem6.click() # driver.find_element_by_id('button-1021-btnIconEl').click() # Click Submit
            stale_button = False
        except:
            stale_button = True
            time.sleep(5)
    
    # Download:
    print(park_chosen+' downloading...')
    stale_button = True
    while (stale_button == True):
        try:
            elem12 = wait.until(ec.element_to_be_clickable((By.ID, "button-1142")))
            elem12.click() # = driver.find_element_by_id('button-1142').click()
            stale_button = False
        except:
            stale_button = True
            time.sleep(5)
    
    driver.close()
```
